```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame
data = {
    "age": [22, 25, 30, 35, 40],
    "income": [20000, 30000, 40000, 45000, 60000],
    "score": [50, 60, 65, 70, 80]
}
df = pd.DataFrame(data)

# Correlation matrix
print(df.corr())

# Heatmap visualization
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```
